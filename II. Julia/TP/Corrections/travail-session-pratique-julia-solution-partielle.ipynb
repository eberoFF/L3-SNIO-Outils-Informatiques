## Ecriture de fonctions

### Ecrire une fonction qui converti les kilomètres en miles

```julia
function ki2mi(km)
    return km / 1.609
end
```

### Ecrire la fonction réciproque

```julia
function mi2ki(mi)
    return mi * 1.609
end
```

### Ecrire une fonction qui converti des litres en gallons (américain)

```julia
function li2ga(li)
    return li / 3.785
end
```

### Ecrire la fonction réciproque

```julia
function ga2li(ga)
    return ga * 3.785
end
```

## Application des fonctions
- Créer un tableau _km_ avec les valeurs de 0 à 100 par pas de 10
- Avec une boucle _for_ afficher la conversion en miles de chacune des valeurs de _km_

```julia
km = collect(0:10:100);
```

```julia
for i in km
    println(i," km ≈ ",round(ki2mi(i),digits=1)," mi")
end
```

- Créer un tableau _mi_ avec les valeurs de 5 à 105 par pas de 5
- Avec une boucle _for_ afficher la conversion en kilomètres de chacune des valeurs de _mi_

```julia
mi2 = collect(5:5:105)
mi = [i for i in 5:5:105];
```

```julia
mi == mi2
```

```julia
for i in mi
    println(i," mi ≈ ",round(mi2ki(i),digits=1)," km")
end
```

- Créer un tableau _li_ avec les valeurs de 1 à 30 par pas de 0.5
- Créer un nouveau tableau _ga_ issu de la conversion en gallons en utilisant les subtilités de julia qui permettent d'appliquer une fonction sur toutes les valeurs d'un tableau (il existe pluseurs solutions possibles - plusieurs réponses seront appréciées)
- Vous pouvez créer autant de cellules que de réponses

```julia
li = collect(1:0.5:30);
```

```julia
ga = broadcast(li2ga,li);
```

Tracez la courbe entre ces deux valeurs avec les différents types de graphiques (lignes, points, barres).

```julia
function convertUE2US(c)
    return ki2mi(100/li2ga(c))
end
```

```julia
convertUE2US(8)
```

```julia
EU = collect(3.5:0.5:18);
```

```julia
US = convertUE2US.(EU);
```

```julia
using Plots
```

```julia
plot(EU,US)
plot!(title="Consommation US/EU)",
    xlabel="Consommation EU (l/100km)",
    ylabel="Consommation US (mpg)",
    legend=false)
scatter!(EU,US)
```
