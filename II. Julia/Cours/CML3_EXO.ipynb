```python
# Lecture du fichier Titanic
df = pd.read_csv("titanic.csv")

selection1 = df[df['Age'] > 30]

selection2 = df[(df['Age'] > 30) & (df['Sex'] == 'male')]

#Sélection de toutes les lignes où l'âge n'est pas supérieur à 30
selection3 = df[~(df['Age'] > 30)]

# Comptage du nombre de lignes dans la DataFrame 'selection'
count1 = len(selection1)
count2 = len(selection2)
count3 = len(selection3)

print("Nombre de lignes où l'âge est supérieur à 30 :", count1)
print("Nombre de lignes où l'âge est supérieur à 30 et le sexe est 'male' :", count2)
print("Nombre de lignes où l'âge n'est pas supérieur à 30 :", count3)
```

```python
# Quel était le taux de survie global des passagers du Titanic ?
import pandas as pd

df = pd.read_csv("titanic.csv")

# Calculer le taux de survie global
taux_de_survie_global = df["Survived"].mean()

# Transformer le taux de survie global en pourcentage avec deux décimales
pourcentage_de_survie_global = taux_de_survie_global * 100

# Formater la chaîne avec deux décimales
pourcentage_formate = "{:.2f}".format(pourcentage_de_survie_global)

print("Taux de survie global en pourcentage :", pourcentage_formate, "%")
```

```python
# Quelle était la répartition des passagers par classe de billet ?

# Calculer la répartition des passagers par classe de billet (Pclass)
repartition_par_classe = df['Pclass'].value_counts()

print("Répartition des passagers par classe de billet :")
print(repartition_par_classe)
```

```python
# Quelle était la répartition des passagers par âge ?

# Définir les intervalles d'âge
intervalle_age = pd.cut(df['Age'], bins=[0, 18, 30, 50, 100], labels=['0-17', '18-29', '30-49', '50+'])

# Calculer la répartition des passagers par âge
repartition_par_age = intervalle_age.value_counts().sort_index()

print("Répartition des passagers par âge :")
print(repartition_par_age)
```

```python
# Quelle était la répartition des passagers par sexe ?

repartition_par_sexe = df['Sex'].value_counts()

print("Répartition des passagers par sexe :")
print(repartition_par_sexe)
```

```python
# Quelle était la moyenne des tarifs payés par les passagers dans chaque classe ?

# Calculer la moyenne des tarifs payés par classe de billet (Pclass)
moyenne_tarifs_par_classe = df.groupby('Pclass')['Fare'].mean()

print("Moyenne des tarifs payés par classe de billet :")
print(moyenne_tarifs_par_classe)
```

```python
# Quels étaient les ports d'embarquement les plus fréquemment utilisés ?

ports_frequents = df['Embarked'].value_counts()

print("Ports d'embarquement les plus fréquemment utilisés :")
print(ports_frequents)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Lecture du fichier titanic.csv
df = pd.read_csv('titanic.csv')

# Définir les catégories d'âge
bins = [0, 18, 30, 50, 100]
labels = ['Enfant', 'Jeune adulte', 'Adulte', 'Senior']

# Ajouter une nouvelle colonne 'Groupe d'âge' basée sur les catégories
df['Groupe d\'âge'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Compter le nombre de passagers par groupe d'âge
age_counts = df['Groupe d\'âge'].value_counts().sort_index()

# Créer un graphique à barres pour représenter la répartition
plt.bar(age_counts.index, age_counts.values)

# Ajouter des étiquettes et un titre au graphique
plt.xlabel('Groupe d\'âge')
plt.ylabel('Nombre de passagers')
plt.title('Répartition des passagers par groupe d\'âge')

# Afficher le graphique
plt.show()
```

```python
# Créer un nuage de points Age vs Fare
plt.scatter(df['Age'], df['Fare'], alpha=0.5)  # alpha ajoute de la transparence aux points
plt.title('Nuage de points Age vs Fare')
plt.xlabel('Âge')
plt.ylabel('Tarif')

# Enregistrer le graphique au format PDF
plt.savefig('graph1.pdf')

# Afficher le graphique
plt.show()
```

```python
# Compter le nombre de personnes ayant survécu et celles n'ayant pas survécu
survived_counts = df['Survived'].value_counts()

# Créer un graphique en secteurs
plt.figure(figsize=(6, 6))
plt.pie(survived_counts, labels=['Non Survécu', 'Survécu'], autopct='%1.2f%%', startangle=180)
plt.title('Répartition des Survivants dans le Titanic')

plt.show()
```

```python
# Création de sous-graphiques (subplots)
# 2 lignes et 2 colonnes de sous-graphiques
# Le premier sous-graphique
plt.subplot(2, 2, 1)
plt.hist(df['Age'].dropna(), bins=20, color='red', alpha=0.7)
plt.title('Distribution des âges')
plt.xlabel('Âge')
plt.ylabel('Fréquence')

# Le deuxième sous-graphique
plt.subplot(2, 2, 2)
plt.bar(df['Pclass'].unique(), df['Pclass'].value_counts().sort_index(), color='green', alpha=0.7)
plt.title('Répartition par classe')
plt.xlabel('Classe')
plt.ylabel('Nombre de passagers')

# Le troisième sous-graphique
plt.subplot(2, 2, 3)
plt.scatter(df['Fare'], df['Age'], color='red', alpha=0.5)
plt.title('Corrélation entre le tarif et l\'âge')
plt.xlabel('Tarif')
plt.ylabel('Âge')

# Le quatrième sous-graphique
plt.subplot(2, 2, 4)
plt.pie(df['Sex'].value_counts(), labels=['Hommes', 'Femmes'], autopct='%1.1f%%', colors=['purple', 'orange'])
plt.title('Répartition par sexe')

# Ajuster la disposition des sous-graphiques
plt.tight_layout()

plt.show()
```

```python
# Créer une figure avec 2x2 sous-graphiques
plt.figure(figsize=(12, 8))

# Créer une liste de couleurs en fonction de la colonne Survived
colors = ['blue' if survived == 1 else 'red' for survived in df['Survived']]

# Sous-plot 1 : Age vs Fare
plt.subplot(2, 2, 1)
plt.scatter(df['Age'], df['Fare'], c=colors, alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare')

# Sous-plot 2 : Age vs SibSp
plt.subplot(2, 2, 2)
plt.scatter(df['Age'], df['SibSp'], c=colors, alpha=0.5)
plt.xlabel('Age')
plt.ylabel('SibSp')
plt.title('Age vs SibSp')

# Sous-plot 3 : Age vs Parch
plt.subplot(2, 2, 3)
plt.scatter(df['Age'], df['Parch'], c=colors, alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Parch')
plt.title('Age vs Parch')

# Sous-plot 4 : Age vs Survived (coloré par Survived)
plt.subplot(2, 2, 4)
plt.scatter(df['Age'], df['Survived'], c=colors, alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Survived')
plt.title('Age vs Survived')

# Ajuster l'espacement entre les sous-graphiques
plt.tight_layout()

plt.show()
```
